```python
import pandas as pd
```

```python
dfFly = pd.read_csv('flights-selected/1912_bts_flights.csv')
```

```python
dfFly.head()
```

```python
y = dfFly['DepDel15']
X = dfFly.drop(['DepDel15','FlightDate','DepDelay','DepDelayMinutes','DepartureDelayGroups','ArrDelay','ArrDelayMinutes','ArrivalDelayGroups',
                'ArrDel15','CarrierDelay','WeatherDelay','NASDelay','SecurityDelay','LateAircraftDelay'], axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

mdlLrFly = LogisticRegression()
mdlLrFly.fit(X_train, y_train)
y_test_pred = mdlLrFly.predict(X_test)
print(classification_report(y_test, y_test_pred))
```
