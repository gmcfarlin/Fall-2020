# Initialize and Load Data

```python
import seaborn as sns
import pandas as pd
```

```python
# Use country as index and state as second index to isolate specific regions easily
dfCovid = pd.read_csv('train.csv', index_col=['Country_Region','Province_State'])

# Convert date into datetime format for plotting
dfCovid['Date'] = pd.to_datetime(dfCovid['Date'])
```

## Visualize Data
source for significant dates: https://www.nytimes.com/article/coronavirus-timeline.html

### New York, US

```python
data = dfCovid.loc['US',:].loc['New York',:]
sns.lineplot(data=data, x='Date', y='ConfirmedCases')

# Feb 28th, US announces travel restrictions
plt.vlines(x=pd.to_datetime('2020-02-28'), ymin=-5, ymax=max(data['ConfirmedCases']), colors='k', label='US travel restrictions')
# March 7th, NY declares state of emergency
plt.vlines(x=pd.to_datetime('2020-03-07'), ymin=-5, ymax=max(data['ConfirmedCases']), colors='r', label='state of emergency')
# March 20th, NY declares shelter-in-place order
plt.vlines(x=pd.to_datetime('2020-03-20'), ymin=-5, ymax=max(data['ConfirmedCases']), colors='g', label='shelter-in-place')
# April 15th, NY requires facemasks
#plt.vlines(x=pd.to_datetime('2020-04-15'), ymin=-5, ymax=max(data['ConfirmedCases']), colors='y')
plt.legend()
```

### Hubei, China

```python
data = dfCovid.loc['China',:].loc['Hubei',:]
sns.lineplot(data=data, x='Date', y='ConfirmedCases')

# Jan 23rd, Chinese government shuts down Wuhan
plt.vlines(x=pd.to_datetime('2020-01-23'), ymin=-5, ymax=max(data['ConfirmedCases']), colors='r', label='shutdown')
# Feb 13th, change in diagnostic criteria (lung scans included) leads to large increase in cases
plt.vlines(x=pd.to_datetime('2020-02-13'), ymin=-5, ymax=max(data['ConfirmedCases']), colors='g', label='change in diagnostic criteria')
# Mar 19th, China reports no new cases in day
plt.vlines(x=pd.to_datetime('2020-03-19'), ymin=-5, ymax=max(data['ConfirmedCases']), colors='m', label='China reports no new cases')
plt.legend()
```

### Italy

```python
data = dfCovid.loc['Italy',:]
sns.lineplot(data=data, x='Date', y='ConfirmedCases')

# Feb 23rd, Italy begins lockdown of small towns
plt.vlines(x=pd.to_datetime('2020-02-23'), ymin=-5, ymax=max(data['ConfirmedCases']), colors='y', label='first lockdown in small towns')
# March 8th, Italy imposes strict quarrantine
plt.vlines(x=pd.to_datetime('2020-03-08'), ymin=-5, ymax=max(data['ConfirmedCases']), colors='r', label='strict quarrantine imposed')
# March 17th, EU orders people not to leave bloc for 30 days
plt.vlines(x=pd.to_datetime('2020-03-17'), ymin=-5, ymax=max(data['ConfirmedCases']), colors='g', label='EU travel ban')
# April 13th, Italy to reopen some shops
plt.vlines(x=pd.to_datetime('2020-04-13'), ymin=-5, ymax=max(data['ConfirmedCases']), colors='m', label='reopening shops')
plt.legend()
```
