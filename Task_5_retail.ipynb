## Create new features for segmenting customers

### Country Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
DfCusLe = DfCus.copy()
DfCusLe['country_name'] = le.fit_transform(DfCus.loc[:,'country_name'])
```

```python
le.classes_
```

### Region Encoding

```python
europe = ['Austria', 'Belgium', 'Czech Republic', 'Denmark', 'Germany', 'Greece', 'Portugal',
          'Poland','Netherlands','Spain', 'Sweden', 'Switzerland', 'United Kingdom', 
          'Channel Islands', 'Italy', 'EIRE', 'Norway', 'Lithuania', 'European Community','France','Finland','Malta']
mid_east = ['Cyprus', 'Israel', 'Saudi Arabia','United Arab Emirates', 'Lebanon', 'Bahrain']
asia = ['Singapore', 'Japan']
north_am = ['USA', 'Canada']
other = ['Unspecified', 'Iceland', 'Brazil', 'Australia', 'RSA']
```

```python
count = 0
for row in DfCusLe.values:
    country = le.inverse_transform([DfCusLe.loc[count, 'country_name']])[0]
    if country in europe:
        region = 'europe'
    elif country in mid_east:
        region = 'mid_east'
    elif country in asia:
        region = 'asia'
    elif country in north_am:
        region = 'north_am'
    elif country in other:
        region = 'other'
    else:
        print(country)
    
    DfCusLe.loc[count, 'region_name'] = region
    count +=1
```

```python
DfCusLe['region_name'] = le.fit_transform(DfCusLe.loc[:,'region_name'])
```

### Create average transaction

```python
count = 0
for row in DfCusLe.values:
    invoices = DfCusLe.loc[count, 'invoice_count']
    total = DfCusLe.loc[count, 'invoice_amt_total']
    DfCusLe.loc[count, 'avg_trans_amt'] = total/invoices
    count +=1
```

### Create average price per item

```python
count = 0
for row in DfCusLe.values:
    quantity = DfCusLe.loc[count, 'quantity_avg']
    trans = DfCusLe.loc[count, 'avg_trans_amt']
    DfCusLe.loc[count, 'avg_price_item'] = trans/quantity
    count +=1
```

## Set target/features

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

def build_score_rfs(X,y):
    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=1)
    Mdl_Cus_Rfs = RandomForestRegressor(random_state=1, n_estimators=100)
    Mdl_Cus_Rfs.fit(X_train, y_train)
    y_test_pred = Mdl_Cus_Rfs.predict(X_test)
    mae = mean_absolute_error(y_test, y_test_pred)
    return mae
```

```python
y = DfCusLe['invoice_amt_total']
X = DfCusLe.drop(['invoice_amt_total', 'region_name', 'avg_price_item', 'avg_trans_amt'], axis=1)
print(build_score_rfs(X,y))
```

```python
y = DfCusLe['invoice_amt_total']
X = DfCusLe.drop('invoice_amt_total', axis=1)
print(build_score_rfs(X,y))
```
